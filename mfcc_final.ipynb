```python
import librosa
import numpy as np
import pandas as pd
from pathlib import Path

```

```python
Data=np.empty([380,1])
i=1
basepath = Path('/home/thirumalakumar/Desktop/Final_Noise/')
files_in_basepath = basepath.iterdir()
for item in files_in_basepath:
    if item.is_file():
            x,fs=librosa.load(item,sr=16000)
            X=librosa.feature.mfcc(x,fs,n_mfcc=13,win_length=int(0.02*fs),hop_length=int(0.01*fs))
            X=np.reshape(X,(X.shape[0]*X.shape[1],1))
            
            X=X[:381,:]
            if item.name[0]=='y':
                X[380]=1
            if item.name[0]=='n':
                X[380]=0
            if i==1:
     
                Data=X
   
            else:
                Data=np.concatenate((Data,X),axis=1)
            i=i+1
            print("iteration:%d"%i,X.shape)

            
            #print(Data.shape)
        
                







```
